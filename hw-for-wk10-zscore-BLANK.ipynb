Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
## import libraries and packages
import pandas as pd

## mad math functions package
from scipy.stats import zscore 


## run the display code here
pd.options.display.float_format = '{:,.2f}'.format
```

```python
## import csv
dflp= pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
dflp
```

```python
## find the mean for reference
round(dflp.groupby("album")["sales"].agg(["mean","std"]),0)
```

```python
##add column for z scores in LPs
dflp["lp_zscore"] = zscore(dflp["sales"])
dflp
```

```python
## Now do the same for albums in 2023

df= pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
df
```

```python
## find the mean for reference for 2023 albums
round(df.groupby("album")["streams_2023"].agg(["mean","std"]),0)
```

```python
##add column for z scores in 2023 year albums (the highest is the furthest away from the mean, upwards = most sold)
df["2023_zscore"] = zscore(df["streams_2023"])
df
```
